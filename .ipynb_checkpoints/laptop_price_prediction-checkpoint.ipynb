### Data cleaning and feature engineering

```python
df1 = df.copy()
```

```python
# Removed the 'Product' column due to its numerous categorical features, which could potentially lead to the curse of dimensionality if we had hot encoded this column.
df1 = df.drop('Product', axis=1)
```

```python
df1.head()
```

```python
# One hot encoding 'Typename' and 'Company' columns
columns_to_OHE = ['TypeName', 'Company', 'OpSys']
for column in columns_to_OHE:
    df1 = df1.join(pd.get_dummies(df[column]))
    df1 = df1.drop(column, axis=1)
    
```

```python
df1.head()
```

```python
# Extract resolution from Screen Resolution and then create new feature, screen_Width and Screen_Height
# Drop ScreenResolution column

df1['ScreenResolution'] = df1.ScreenResolution.str.split(' ').apply(lambda x : x[-1])
df1['Screen_Width'] = df1['ScreenResolution'].str.split('x').apply(lambda x: x[0])
df1['Screen_Width'] = df1['Screen_Width'].astype('int')
df1['Screen_Height'] = df1['ScreenResolution'].str.split('x').apply(lambda x: x[-1])
df1['Screen_Height'] = df1['Screen_Height'].astype('int')
```

```python
df1.drop('ScreenResolution', axis=1, inplace=True)
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
# Extract Cpu name and frequency from 'Cpu' column and make new columns 'Cpu_Brand' and 'Cpu_frequency'
# Drop 'Cpu' column

df1['Cpu_brand'] = df1['Cpu'].str.split(' ').apply(lambda x: x[0])
df1['Cpu_frequency'] = df1['Cpu'].str.split(' ').apply(lambda x: x[-1][:-3])
```

```python
df1 = df1.drop('Cpu', axis=1)
```

```python
df1['Cpu_frequency'] = df1['Cpu_frequency'].astype('float')
```

```python
# Remove GB from the end for the column 'Ram' - We will convert all the Object columns to int/Float later
df1['Ram'] = df1['Ram'].str[:-2]
```

```python
df1['Ram'] = df1['Ram'].astype('int')
```

```python
# df1.dtypes
```

```python
df1['Memory_size'] = df1['Memory'].str.split(' ').apply(lambda x: x[0])
df1['Memory_type'] = df1['Memory'].str.split(' ').apply(lambda x: x[-1])
```

```python
df1.head()
```

```python
def clean(value):
    if 'GB' in value:
        return float(value[:value.find('GB')]) * 1000
    elif 'TB' in value:
        return float(value[:value.find('TB')]) * 100000
```

```python
df1['Memory_size'] = df1['Memory_size'].apply(clean)
```

```python
df1 = df1.drop('Memory', axis=1)
```

```python
df1.head()
```

```python
# Remove kg from the weight column
df1['Weight'] = df1['Weight'].str[:-2]
```

```python
df1['Weight'] = df1['Weight'].astype('float')
```

```python
df1['Gpu_name'] = df1['Gpu'].str.split(' ').apply(lambda x: x[0])
```

```python
df1 = df1.drop('Gpu', axis=1)
```

```python
df1.head()
```

```python
df1 = df1.join(pd.get_dummies(df1['Memory_type']))
```

```python
df1 = df1.drop('Memory_type', axis=1)
df1.head()
```

```python
cpu_categories = pd.get_dummies(df1['Cpu_brand'])

cpu_categories.columns = [col + '_cpu' for col in cpu_categories.columns]
```

```python
df1 = df1.join(cpu_categories)
df1 = df1.drop('Cpu_brand', axis=1)
```

```python
gpu_categories = pd.get_dummies(df1['Gpu_name'])
gpu_categories.columns = [col + '_gpu' for col in gpu_categories.columns]
```

```python
df1 = df1.join(gpu_categories)
df1 = df1.drop('Gpu_name', axis=1)
```

```python
df1.head()
```

```python
df1.head()
```

```python
correlations = df1.corr()['Price_euros'].apply(abs).sort_values(ascending=False)
correlations
```

```python
selected_columns = correlations[correlations>0.1].index
```

```python
final_df = df1[selected_columns]
final_df.head()
```

```python
plt.figure(figsize=(18,15))
sns.heatmap(final_df.corr(), annot=True, cmap='YlGnBu')
```

### Modelling and Prediction

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
```

```python
X,y = final_df.drop('Price_euros', axis=1), final_df['Price_euros']
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.85)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train_scaled, y_train)
```

```python
model.score(X_test_scaled, y_test)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_pred, y_test)
plt.plot(range(0,3500), range(0,3500), c='red')
```
